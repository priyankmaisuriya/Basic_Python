```
#Python Program to check if two Strings are Anagram.

s1 = sorted('race')
s2 = ['care']

for w in s2:
    if s1 == sorted(w):
        print("Yes,Anagram:", w)
    else:
        print("Not,Anagram:", w)
```

```
#Python program to convert lowercase vowel to uppercase in string.

string="Rocky is a Gangstar"
kgf = ""
for w in string:
    if w in ['a', 'e', 'i', 'o', 'u'] or w in ['A','E','I','O','U']:
        kgf += w.lower()
    else:
        kgf += w.upper()
print(kgf)
```

```
#Python program to count Occurrence Of Vowels & Consonants in a String.

string="rocky bhai"
vowels=0
consonants=0
for i in string:
    if i in ('a,e,i,o,u'):
        vowels+=1
    elif i.isalpha():
        consonants+=1
print("Vowels :",vowels,"Consonants:",consonants)
```

```
#Python program to count alphabets, digits and special characters.

string="rocky @#$ dinanath_ m1111"
alpha=digit=special=0

for i in range(len(string)):
    if((string[i] >= 'a' and string[i] <= 'z') or (string[i] >= 'A' and string[i] <= 'Z') ):
        alpha = alpha + 1
    elif(string[i] >= '0' and string[i] <= '9'):
        digit = digit + 1
    else:
        special = special + 1
print("Alphabets:",alpha)
print("Digits:",digit)
print("Special Characters:",special)
```
