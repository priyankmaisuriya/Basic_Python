```
#Find the Max of three numbers
def max_two(x,y):
    if x>y:
        return x
    return y
#print(max_two(4,8))
def max_three(x,y,z):
    return max_two(x,max_two(y,z))
print(max_three(4,8,-10))
```

```
#Python function to sum all the numbers in a list.
def sum(num):
    total=0
    for x in num:
        total+=x
    return total
print(sum((5,6,11,10,2)))
```

```
#Python function to multiply all the numbers in a list.
def multi(num):
    total=1
    for x in num:
        total*=x
    return total
print(multi((5,6,11,10,2)))
```

```
#Arbitrary Arguments, *args
def child(*kids):
    print("youngest child is " +kids[2])
child("meet","vijay","vicky")
```

```
#Arbitrary Keyword Arguments, **kwargs
def sur(**lname):
    print("his last name is " + lname["lastname"])
sur(firstname="rocky",lastname="thalapati")
```

```
#Default Parameter Value
def email(key="@"):
    print("vijaythalapati"+key+"gmail.com")
email()
```

```
#Add 10 to argument a, and return the result
x=lambda a:a+10
print(x(5))
```

```
#Multiply and sum argument a with argument b and return the result
x=lambda a,b,c:a * b + c
print(x(5,7,10))
```

```
#doubles the number
def num(n):
    return lambda a:a * n
double = num(2)
print(double(22))
```

```
#itertor top ten
class topten:
    
    def __init__(self):
        self.num=1
    def __iter__(self):
        return self
    def __next__(self):
        if self.num<=10:
            val=self.num
            self.num +=1
            return val
        else:
            raise StopIteration
values = topten()
#print(next(values))
for i in values:
    print(i)
```
