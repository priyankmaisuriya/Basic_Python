```
#file header name display
sdf = spark.read.format("csv").option("header", "true") .option("inferSchema", "true") .load('/FileStore/tables/file.csv')
display(sdf)
```

```
#display one or more column
sdf.select("id","first_name","last_name").display()
```

```
#data display
sdf.select("first_name",sdf.last_name.like("patel")).display()
```

```
#row data display
sdf.select("first_name","city",sdf.last_name.like("patel")).head()
```

```
#one and more row data display
sdf.select("first_name","city",sdf.last_name.like("patel")).head(3)
```

```
#first row data display
sdf.first()
```
